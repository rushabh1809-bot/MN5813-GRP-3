```python
import pandas as pd
import requests
from zipfile import ZipFile
import io
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://raw.githubusercontent.com/rushabh1809-bot/MN5813-GRP-3/main/athlete_eventsgrp.csv.zip'

retrived_url = requests.get(url)
zip_file = ZipFile(io.BytesIO(retrived_url.content))

df= pd.read_csv(zip_file.open('athlete_events.csv'))
df.head()


```

The data set of the olympic data has been loaded in from a zipcode in Github and 5 rows has been shown of the data frame to ensure it has loaded in

```python
print('What columns are in our data set:')
print(df.columns)
```

```python
duplicated_data=df.duplicated().sum()
print(f'Amount of duplicates in data set: {duplicated_data}')


```

```python
df=df.drop_duplicates()
```

Duplicates are deleted


```python
df['Height']=df['Height'].fillna(df['Height'].median())
df['Weight']=df['Weight'].fillna(df['Weight'].median())
df['Age']=df['Age'].fillna(df['Age'].median())
df['Medal']=df['Medal'].fillna('No Medal')
```

Nan values have been changed into median of data so data can be easily analysed without it being skewed, we assumed nan values for medals meant that the athlete didnt recieve a medal. 

```python
medal_table = pd.pivot_table(
    df,
    index='NOC',     
    columns='Year',     
    values='Medal',     
    aggfunc='count',      
    fill_value=0         
)

medal_table
cleaned_table = (
    medal_table
    .assign(Total=medal_table.sum(axis=1))
    .sort_values("Total", ascending=False)
    .head(20)                      
)
cleaned_table
```

```python
noc_cleaned_zeros = cleaned_table[(cleaned_table == 0).sum(axis=1) > 10]
noc_cleaned_zeros
```

These are the top 20 regions who won the most medals in ascending order, America seems to have dominated the Olympics over the 120 years winning over 6000 more medals than the second highest region which is France. In particular, what is interesting is how three regions: URS, RUS, CHN have achieved the top 20 medals won despite having over 10 years of achieving 0 medals in the Olympics. It is worth highlighting the history of the region of URS as the soviet union ran in 1952-1988 and was discontinued which corresponds with the data shown.

```python
df['Year']= pd.to_datetime(df['Year'], format= '%Y')
df['Century']= (df['Year'].dt.year//100)+1
```

```python
bins=[0,20,40,60,80,100]
labels= ['0-20', '20-40', '40-60', '60-80','80-100']
df['Age group']= pd.cut(df['Age'], bins=bins, labels=labels,right=False)
df
```

```python

# Sort average age
average_age_event = df.groupby('Event')['Age'].mean().astype(int).sort_values()

plt.figure(figsize=(12,8))
# Color gradient based on age
colors = plt.cm.coolwarm((average_age_event - average_age_event.min()) / (average_age_event.max() - average_age_event.min()))
plt.barh(average_age_event.index, average_age_event.values, color=colors)
plt.xlabel('Average Age', fontsize=14)
plt.ylabel('Event', fontsize=14)
plt.title('Average Age per Event', fontsize=16)
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()

```

```python
gold_medal=df[df['Medal']=='Gold']
top_10=gold_medal.groupby('NOC')['Medal'].count().sort_values(ascending=False).head(10)
print(top_10)
```

```python
medal_count = df.groupby(['Team', 'Medal']).size().unstack(fill_value=0)

for medal in ["Gold", "Silver", "Bronze"]:
    if medal not in medal_count.columns:
        medal_count[medal] = 0

medal_count["Total"] = medal_count["Gold"] + medal_count["Silver"] + medal_count["Bronze"]

top5 = medal_count.sort_values("Total", ascending=False).head(5)

plt.figure(figsize=(10, 6))

plt.bar(top5.index, top5["Gold"], label="Gold", color="#FFD700")
plt.bar(top5.index, top5["Silver"], bottom=top5["Gold"], label="Silver", color="#C0C0C0")
plt.bar(top5.index, top5["Bronze"], bottom=top5["Gold"] + top5["Silver"], label="Bronze", color="#CD7F30")

plt.title("Medal Distribution of Top 5 Countries")
plt.xlabel("Country")
plt.ylabel("Number of Medals")
plt.legend(title="Medal Type")

plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
winter_awards = df[(df['Medal'].notna()) & (df['Season'] == 'Winter')].groupby('Year')['Medal'].count()
summer_awards = df[(df['Medal'].notna()) & (df['Season'] == 'Summer')].groupby('Year')['Medal'].count()

winter_smooth = winter_awards.rolling(window=2, min_periods=1).mean()
summer_smooth = summer_awards.rolling(window=2, min_periods=1).mean()

# Plot smooth lines
plt.figure(figsize=(12,6))
plt.plot(winter_smooth.index, winter_smooth.values, color='blue', linewidth=2, label='Winter')
plt.plot(summer_smooth.index, summer_smooth.values, color='red', linewidth=2, label='Summer')

plt.title("Medals by Year: Winter vs Summer Olympics", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Medals", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)
plt.show()
```

```python

top_athletes = df.groupby(['Name','Sport'])['Medal'].count().sort_values(ascending=False).head(10)

y_labels = [f"{name}\n{sport}" for name, sport in top_athletes.index]

plt.figure(figsize=(10,6))
plt.hlines(y=y_labels, xmin=0, xmax=top_athletes.values, color='skyblue', alpha=0.7)
plt.plot(top_athletes.values, y_labels, "o", color='blue')
plt.xlabel("Number of Medals")
plt.title("Top 10 Athletes by Medals")
plt.gca().invert_yaxis()  
plt.show()

```
