```python
import pandas as pd
import requests
from zipfile import ZipFile
import io

url = 'https://raw.githubusercontent.com/rushabh1809-bot/MN5813-GRP-3/main/athlete_events.csv.zip'

retrived_url = requests.get(url)
zip_file = ZipFile(io.BytesIO(retrived_url.content))

df= pd.read_csv(zip_file.open('athlete_events.csv'))
df.head()


```

```python
df_duplicates= df[df.duplicated(keep=False)]
no_duplicates= df.drop(df_duplicates.index,inplace=True)
df.columns= df.columns.str.lower()
```

```python
df['height'].fillna(df['height'].median(),inplace=True)
df['weight'].fillna(df['weight'].median(),inplace=True)
df['age'].fillna(df['age'].median(),inplace=True)
df['medal'].fillna('No Medal',inplace=True)
```
